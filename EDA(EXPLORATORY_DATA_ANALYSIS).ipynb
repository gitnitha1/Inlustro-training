```python
#scatter plot
import seaborn as sns
arr_text=pd.read_csv('/content/train.csv')
df=pd.DataFrame(arr_text)
plt.figure(figsize=(10,7))
sns.histplot(df['Fare'],bins=20,kde=True,color="blue")
plt.title("HISTOGRAM PLOT",fontsize=30,fontstyle="italic")
plt.show()
```

```python
#barplot
plt.figure(figsize=(10,7))
sns.barplot(x='Ticket',y='Fare',data=arr_text,palette="coolwarm")
plt.title("BARPLOT",fontsize=20,fontstyle='oblique')
plt.show()
```

```python
#bOXplot
plt.figure(figsize=(10,7))
sns.boxplot(x='Ticket',y='Fare',data=arr_text,palette="coolwarm")
plt.title("BoxPLOT",fontsize=20,fontstyle='oblique')
plt.show()
```

```python
#SCATTERPLOT
plt.figure(figsize=(10,7))
sns.scatterplot(x='Ticket',y='Fare',data=arr_text,palette="coolwarm")
plt.title("SCATTERPLOT",fontsize=20,fontstyle='oblique')
plt.show()
```
